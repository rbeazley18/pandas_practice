```python
location_col = ['Location']

def strip_location(current_val):
    before, sep, after = current_val.partition('\ufeff')
    new_val = before
    return new_val 

def clean_location(row):
    for col in location_col:
        current_val = row[col]
        row[col] = strip_location(current_val)
    return row

df = df.apply(clean_location, axis=1)        
```

```python
location_col = ['City']

def partition_city(current_val):
    before, sep, after = current_val.partition('[')
    new_val = before
    return new_val 

def clean_city(row):
    for col in location_col:
        current_val = row[col]
        row[col] = partition_city(current_val)
    return row

df = df.apply(clean_city, axis=1)
```

```python
old_col = ['2020 land area']

def partition_land_area(current_val):
    before, sep, after = current_val.partition('sq')
    new_val = before.rstrip()
    return new_val 

def clean_land_area(row):
    for col in old_col:
        current_val = row[col]
        row[col] = partition_land_area(current_val)
    return row

df = df.apply(clean_land_area, axis=1)
```

```python
df.rename(columns={'2020 land area': '2020 land area (square miles)', '2020 population density': '2020 population density (people per square mile)'}, inplace=True)
```

```python
old_col = ['2020 population density (people per square mile)']

def partition_pop_density(current_val):
    before, sep, after = current_val.partition('/')
    new_val = before
    return new_val 

def clean_pop_density(row):
    for col in old_col:
        current_val = row[col]
        row[col] = partition_pop_density(current_val)
    return row

df = df.apply(clean_pop_density, axis=1)
```
