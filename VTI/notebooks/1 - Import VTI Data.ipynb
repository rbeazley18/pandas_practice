```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
cwd = os.getcwd()
print(cwd)
BASE_DIR = os.path.dirname(cwd)
print(BASE_DIR)
DATA_DIR = os.path.join(BASE_DIR, 'data')
print(DATA_DIR)
```

```python
os.listdir(DATA_DIR)
```

```python
VTI_data = os.path.join(DATA_DIR, 'VTI_data.csv')
print(os.path.exists(VTI_data))
```

```python
d_parser = lambda x: datetime.strptime(x, '%m/%d/%Y')
# df = pd.read_csv('ETH_1h.csv', parse_dates=['Date'], date_parser=d_parser)
df = pd.read_csv(VTI_data, parse_dates=['Date'], date_parser=d_parser)
```

```python
df.set_index('Date', inplace=True)
```

```python
df
```

```python
df.rename(columns={'Close/Last': 'Close'}, inplace=True)
df
```

```python
df['High']
```

```python
weekly_highs = df['High'].resample('W').max()
weekly_lows = df['Low'].resample('W').min()
weekly_max_diff = (weekly_highs - weekly_lows).resample('W').max()
```

```python
plt.plot(weekly_highs, label='Weekly High')
plt.plot(weekly_lows, label='Weekly Low')
plt.plot(weekly_max_diff, label='Max Weekly Difference')
plt.legend(loc="upper left")
plt.title("VTI Weekly Max High & Min Low")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.show()
```
